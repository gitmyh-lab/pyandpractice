```python
import matplotlib.pyplot as plt
#生成从-20到20，元素数为10的等差数列
x=np.linspace(-20,20,10)
y=x**3+2*x**2+6*x+5
plt.plot(x,y,marker='o')
plt.show()#其实不用这一句也可以
```

```python
#K最邻近算法分类
#导入数据集生成器
from sklearn.datasets import make_blobs
#导入KNN分类器
from sklearn.neighbors import KNeighborsClassifier
#导入画图工具
import matplotlib.pyplot as plt
#导入数据集拆分工具
from sklearn.model_selection import train_test_split
#生成样本数为200，分类为2的数据集
data=make_blobs(n_samples=200,centers=2,random_state=8)
X,y=data
#将生成的数据集进行可视化
plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.spring,edgecolor='k')
```

```python
#K最邻近算法拟合数据
import numpy as np
clf=KNeighborsClassifier()
clf.fit(X,y)
#画图
x_min,x_max=X[:,0].min()-1,X[:,0].max()+1
y_min,y_max=X[:,1].min()-1,X[:,1].max()+1
xx,yy=np.meshgrid(np.arange(x_min,x_max,.02),
                 np.arange(y_min,y_max,.02))
Z=clf.predict(np.c_[xx.ravel(),yy.ravel()])
Z=Z.reshape(xx.shape)
plt.pcolormesh(xx,yy,Z,cmap=plt.cm.Pastel1)#注意这里是Pastel1 ：L一
plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.spring,edgecolor='k')
plt.xlim(xx.min(),xx.max())
plt.ylim(yy.min(),yy.max())
plt.title('Classifier:KNN')
#把新的数据点用五星表示出来
plt.scatter(6.75,4.82, marker='*', c='red', s=200)
```

```python
#对新数据点分类进行判断
print('\n\n\n')#这一行代码主要是为了让截图好看一些
print('代码运行结果：')
print('==============================')#打印分隔符让结果美观一些
print('新数据点的分类是：',clf.predict([[6.75,4.82]]))
print('==============================')#同上
print('\n\n\n') #这一行代码主要是为了让截图好看一些
```

```python
#生成样本数为500，分类为5的数据集
data2=make_blobs(n_samples=500,centers=5,random_state=8)
X2,y2=data2
#将生成的数据集进行可视化
plt.scatter(X2[:,0],X2[:,1],c=y2,cmap=plt.cm.spring,edgecolor='k')
```

```python
clf=KNeighborsClassifier()
clf.fit(X2,y2)
#画图
x_min,x_max=X2[:,0].min()-1,X2[:,0].max()+1
y_min,y_max=X2[:,1].min()-1,X2[:,1].max()+1
xx,yy=np.meshgrid(np.arange(x_min,x_max,.02),
                 np.arange(y_min,y_max,.02))
Z=clf.predict(np.c_[xx.ravel(),yy.ravel()])
Z=Z.reshape(xx.shape)
plt.pcolormesh(xx,yy,Z,cmap=plt.cm.Pastel1)#注意这里是Pastel1 ：L一
plt.scatter(X2[:,0],X2[:,1],c=y2,cmap=plt.cm.spring,edgecolor='k')
plt.xlim(xx.min(),xx.max())
plt.ylim(yy.min(),yy.max())
plt.title('Classifier:KNN')
```

```python
#将模型的评分进行打印
print('\n\n\n')
print('代码运行结果：')
print('==============================')
print('模型正确率：{:.2f}'.format(clf.score(X2,y2)))
print('==============================')
print('\n\n\n')
```

```python
#K最近邻算法回归
from sklearn.datasets import make_regression
#生成特征数为1，噪音为50的数据集
X,y=make_regression(n_features=1,n_informative=1,noise=50,random_state=8)#标准差为50的noise
#散点图数据可视化
#横轴代表的是样本特征的数值，范围大概在-3～3；纵轴代表样本的测定值，范围大致在-250～250。
plt.scatter(X,y,c='orange',edgecolor='k')
```

```python
#导入用于回归分析的KNN模型
from sklearn.neighbors import KNeighborsRegressor
reg=KNeighborsRegressor()
#用KNN模型拟合数据
reg.fit(X,y)
#预测结果可视化
z=np.linspace(-3,3,200).reshape(-1,1)
plt.scatter(X,y,c='orange',edgecolor='k')
plt.plot(z,reg.predict(z),c='k',linewidth=3)
plt.title('KNN Regression')
```

```python
print('\n\n\n')
print('代码运行结果：')
print('==============================')
print('模型评分：{:.2f}'.format(reg.score(X,y)))
print('==============================')
print('\n\n\n')
```

```python
from sklearn.neighbors import KNeighborsRegressor
reg2=KNeighborsRegressor(n_neighbors=2)
#用KNN模型拟合数据
reg2.fit(X,y)
#预测结果可视化
z=np.linspace(-3,3,200).reshape(-1,1)
plt.scatter(X,y,c='orange',edgecolor='k')
plt.plot(z,reg2.predict(z),c='k',linewidth=3)
plt.title('KNN Regression')
```

```python
print('\n\n\n')
print('代码运行结果：')
print('==============================')
print('模型评分：{:.2f}'.format(reg2.score(X,y)))
print('==============================')
print('\n\n\n')
```
