```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,10,0.1)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,x,y2,scalex=True,scaley=True,color='blue')
plt.plot(2*x,2*y1,2*x,2*y2,scalex=True,scaley=True,color='black')
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,10,0.1)
y1=np.sin(x)
y2=np.cos(x)
dict_1={'xValue':x,'sin':y1,'cos':y2,'scalex':True,'scaley':True}
dict_2={'xValue':2*x,'sin':2*y1,'cos':2*y2,'scalex':True,'scaley':True}
plt.plot('xValue','sin',scalex='scalex',scaley='scaley',color='blue',data=dict_1)
plt.plot('xValue','cos',scalex='scalex',scaley='scaley',color='blue',data=dict_1)
plt.plot('xValue','sin',scalex='scalex',scaley='scaley',color='black',data=dict_2)
plt.plot('xValue','cos',scalex='scalex',scaley='scaley',color='black',data=dict_2)
```

```python
x=np.linspace(-np.pi,np.pi)
plt.plot(x,np.cos(x),x,np.sin(x))
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],['$-\pi$','$-\pi/2$','$0$','$\pi/2$','$\pi$'])
plt.yticks([-1,0,1])
ax=plt.gca()#xy
#ax.spines
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
#plt.legend()
plt.legend(['cosine','sine'],loc='upper left',frameon=False)
#plt.text()
plt.text(np.pi/4-0.15,-0.17,r'$\pi/4$')
plt.plot([np.pi/4,np.pi/4],[0,np.cos(np.pi/4)],'--k')
plt.text(-0.6,np.cos(np.pi/4)-0.05,r'$\sqrt{2}/2$')
plt.plot([0,np.pi/4],[np.cos(np.pi/4),np.cos(np.pi/4)],'k--')
#plt.fill_between()x
plt.fill_between(np.linspace(0,np.pi/4),np.sin(np.linspace(0,np.pi/4)))
plt.fill_between(np.linspace(np.pi/4,np.pi/2),np.cos(np.linspace(np.pi/4,np.pi/2)))
```

```python
import numpy as np
xAxis = np.linspace(-np.pi,np.pi,200,endpoint=True)
ycos,ysin = np.cos(xAxis),np.sin(xAxis)
fig = plt.figure(figsize=(10,6),dpi = 80)
ax = fig.add_subplot(111)
ax.grid(True,linestyle='-.',color='r')
#
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
#
ax.xaxis.set_ticks_position('bottom')
ax.spines['left'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
#
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(15)  #  
    label.set_bbox(dict(facecolor='white',edgecolor='none',alpha=0.60))
#
ax.set_xbound(xAxis.min()*1.1,xAxis.max()*1.1)
ax.set_ybound(ycos.min()*1.1,ycos.max()*1.1)
#
ax.set_xticks((-np.pi,-np.pi/2,np.pi/2,np.pi))
ax.set_yticks([-1,-0.5,0,0.5,1])
ax.plot(xAxis,ycos,color='red',linewidth=2.0,label='cos')
ax.plot(xAxis,ysin,color='blue',linewidth=2.0,label='sin')
ax.legend(loc='upper right',title='')
```
