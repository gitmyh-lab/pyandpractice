```python
x=np.linspace(-np.pi,np.pi)
plt.plot(x,np.cos(x),x,np.sin(x))
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],['$-\pi$','$-\pi/2$','$0$','$\pi/2$','$\pi$'])
plt.yticks([-1,0,1])
ax=plt.gca()#让x轴与y轴在原点相交
#四个坐标轴（上下左右）可以通过ax.spines属性索引得到。考虑到刻度在左边与下方的两个坐标轴上，因此，可以利用索引先将上面的轴和右边的轴设为透明：
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#再将左边与下方两个坐标轴的位置设在数据点的原点：
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
#使用plt.legend()函数来加入图例
plt.legend(['cosine','sine'],loc='upper left',frameon=False)
#利用plt.text()函数，可以向图像中添加两条曲线在第一象限中的交点信息
plt.text(np.pi/4-0.15,-0.17,r'$\pi/4$')
plt.plot([np.pi/4,np.pi/4],[0,np.cos(np.pi/4)],'--k')
plt.text(-0.6,np.cos(np.pi/4)-0.05,r'$\sqrt{2}/2$')
plt.plot([0,np.pi/4],[np.cos(np.pi/4),np.cos(np.pi/4)],'k--')
#可以用plt.fill_between()函数将两条曲线与x轴围成区域上色
plt.fill_between(np.linspace(0,np.pi/4),np.sin(np.linspace(0,np.pi/4)))
plt.fill_between(np.linspace(np.pi/4,np.pi/2),np.cos(np.linspace(np.pi/4,np.pi/2)))
```

```python
import matplotlib.pyplot as plt
#正常显示中文
plt.rcParams['font.sans-serif'] = ['SimHei']
#正常显示负号
plt.rcParams['axes.unicode_minus'] = False
#创建画布
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
ax.grid(True,linestyle='-',color='r')
monthlist = ['1月','2月','3月','4月','5月','6月']
dmdSales = [13,10,27,33,30,45]
pltSales = [1,10,7,26,20,25]
ax.plot(monthlist,dmdSales,'-xb',label='钻石')
ax.plot(monthlist,pltSales,'--dr',label='铂金')
ax.set_xlabel('月份')
ax.set_ylabel('每月销量')
ax.set_title('珠宝销售图')
ax.legend(loc='upper left',title='提示')
```

```python
import numpy as np
xAxis = np.linspace(-np.pi,np.pi,200,endpoint=True)
ycos,ysin = np.cos(xAxis),np.sin(xAxis)
fig = plt.figure(figsize=(10,6),dpi = 80)
ax = fig.add_subplot(111)
ax.grid(True,linestyle='-.',color='r')
#上方和右边坐标轴颜色设为无
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
#设置坐标轴刻度线显示位置
ax.xaxis.set_ticks_position('bottom')
ax.spines['left'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
#设置坐标刻度的字体参数，增加半透明背景
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(15)  #坐标刻度大小 坐标背景 
    label.set_bbox(dict(facecolor='white',edgecolor='none',alpha=0.60))
#设置横纵坐标轴的刻度范围
ax.set_xbound(xAxis.min()*1.1,xAxis.max()*1.1)
ax.set_ybound(ycos.min()*1.1,ycos.max()*1.1)
#设置横纵坐标轴的刻度和标签
ax.set_xticks((-np.pi,-np.pi/2,np.pi/2,np.pi))
ax.set_yticks([-1,-0.5,0,0.5,1])
ax.plot(xAxis,ycos,color='red',linewidth=2.0,label='cos余弦函数')
ax.plot(xAxis,ysin,color='blue',linewidth=2.0,label='sin正弦函数')
ax.legend(loc='upper right',title='提示')
```
